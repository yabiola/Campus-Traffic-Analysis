## Traffic Data 
### Post-Processing of a csv file to increase efficiency in Data Entry
- This python notebook is used to post process data entry on a csv file. The data was collected on a paper form and was manually entered into a csv file.
- I identified that there are some constant values in the form that could be recored easily using python which are:

   - **'Date'**: the date the survey was conducted
   - **'Direction'**: the direction of traffic that was recorded
   - **'Time'**: the time the vehicle was recored to the nearest 5 mins (there were muliply values for the same time)

```python
import pandas as pd # importing the pandas library
```

```python
traffic_data = pd.read_csv('raw_data.csv')  #read the csv file into a pandas dataframe
traffic_data.head() #display the first 5 rows of the dataframe
```

```python
#insert the constant date value '25-10-2022' into a new column 'Date' and position the column as the first column
traffic_data.insert(0, 'Date', '25-10-2022' )

#insert an empty column 'Time' and position the column as the 2nd column
traffic_data.insert(1, 'Time', '')

#insert the constant value 'out' into a new column 'Direction' and position the column as the 3rd column
traffic_data.insert(2, 'Direction', 'out')
```

```python
range(len(traffic_data)) #checking for the range of the length of rows in the dataframe
```

```python
#create an empty list then append and assign the values of the list using the conditions of the index position
time = []
for i in range(len(traffic_data)): #for loop iterates between the length of the rows in the dataset
    if i < 4: time.append('10.30am') # create the conditions that will be used to assign values to the column
    elif i < 14: time.append('10.35am')
    elif i < 27: time.append('10.40am')
    elif i < 31: time.append('10.45am')
    elif i < 38: time.append('10.50am')
    elif i == 38: time.append('10.55am')
    elif i < 50: time.append('11.00am')
    else: break # end when the conditions are not met
```

```python
traffic_data['Time'] = time #convert the 'time' list into the column 'Time' in the dataframe
traffic_data
```

```python
traffic_data.to_csv('complete_data.csv', index = False) #save the dataframe to a csv file format
```
